```python
import pandas as pd

data = {
    "Name": ["Neila", "Madina", "Zarina", "Maya", "Tolganay"],
    "Age": [19, 28, 30, 25, 18],
    "Score": [90, 69, 100, 95, 92]
}

df = pd.DataFrame(data)
df
```

```python
print("First 3 rows:")
print(df.head(3))

print("\nSummary statistics:")
print(df.describe())

avg_score = df["Score"].mean()
print("\nAverage score:", avg_score)
```

```python
older = df[df["Age"] > 20]
print("Students older than 20:")
print(older)

above = df[df["Score"] > 80]
print("\nStudents with score above 80:")
print(above)
```
