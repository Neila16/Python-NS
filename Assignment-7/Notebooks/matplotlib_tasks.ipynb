```python
import matplotlib.pyplot as plt

names = ["Neila", "Zarina", "Madina", "Maya", "Tolganay"]
scores = [85, 78, 90, 88, 92]
```

```python
plt.figure()
plt.plot(names, scores, marker='o', linestyle='-', color='blue')
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Line Plot")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.bar(names, scores, color='green')
plt.xlabel("Students")
plt.ylabel("Scores")
plt.title("Student Scores Bar Chart")
plt.show()
```
